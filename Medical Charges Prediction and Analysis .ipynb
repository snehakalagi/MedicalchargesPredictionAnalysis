# Medical Charges Prediction and Analysis 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
mcp = pd.read_csv("data.csv")
```

```python
#Age Distribution
sns.set()
plt.figure(figsize=(6,6))
sns.histplot(mcp['age'])
plt.title('1. Age Distribution')
plt.show()
```

```python
#Distribution of Various Properties based on Gender
plt.figure(figsize=(6,6))
sns.countplot(x='sex',data=mcp)
plt.title('2.Gender Graph')
plt.show()
```

```python
mcp['sex'].value_counts()
```

```python
#Perform hypothesis testing to check whether having children more children results in the fact that people smoke less?
plt.figure(figsize=(6,6))
sns.countplot(x='children',data=mcp)
plt.title('3.Children Graph')
plt.show()
```

```python
# Age vs Charges
plt.figure(figsize=(6,6))
sns.scatterplot(data=mcp, x="age", y="charges")#scatterplot(x='age', y='charges', c='region', colormap='viridis')
plt.title('4.Age vs charges graph')
plt.show()
```

```python
# BMI vs Charges -> smoker
plt.figure(figsize=(6,6))
sns.scatterplot(data=mcp, x="bmi", y="charges" , hue="smoker") 
plt.title('BMI vs charges graph')
plt.show()
```

```python
# Children vs charges
plt.figure(figsize=(6,6))
sns.scatterplot(data=mcp, x="children", y="charges")
plt.title('6.children vs charges graph')
plt.show()
```

Observation
1. "children" does not affect the charge significantly
2. The above "BMI vs the charges graph" shows that both "bmi" and "smoker" together form a synergistic effect on charge
3. "age" column also influences the charge

```python
#What can you say about the difference in distribution of charges for smokers and non smokers?
#Smokers vs charges
plt.figure(figsize=(6,6))
sns.scatterplot(data=mcp, x="smoker", y="charges")
plt.title('smoker vs charges graph')
plt.show()
```

```python
Observation
From the above graph we can see that "smoker" independently is not affecting the "charges"
```

```python
#How is BMI affecting the distribution of charges.
# BMI vs Charges  -> Age
plt.figure(figsize=(6,6))
sns.scatterplot(data=mcp, x="bmi", y="charges", hue="age" ) 
plt.title('BMI vs charges graph')
plt.show()
```
